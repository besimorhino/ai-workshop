# Glossary of common AI/ML terms
Here’s a glossary of common AI terms with easy to understand (but not insulting your intelligence) definitions.

**AI (Artificial Intelligence)**: The ability of machines to mimic human thinking and perform tasks like learning, problem-solving, or recognizing patterns.

**AI Agent**: A computer program that can make decisions and take actions on its own to solve tasks or achieve goals.

**Accuracy**: The percentage of correct predictions made by a model out of all predictions.

**Activation Function**: A mathematical function used in neural networks that determines whether a neuron should be activated or not, helping the network capture complex patterns in data.

**Algorithm**: A step-by-step set of instructions for solving a problem or performing a task.

**Backpropagation**: The process of updating weights in a neural network by calculating the error (difference between predicted and actual output) and adjusting the weights accordingly.

**Batch Size**: The number of training examples used in one iteration of model training before updating weights.

**Bias**: When an AI model favors certain outcomes over others due to imbalanced or flawed training data.

**Computer Vision**: A field of AI that enables computers to interpret and process visual data, like images and videos.

**Confusion Matrix**: A table that shows the performance of a model by comparing predicted and actual outcomes.

**Corpus**: A large collection of text or documents used to train or evaluate AI models, especially in natural language processing tasks.

**Corpus Poisoning**: A type of attack where malicious or manipulated data is intentionally introduced into the training dataset (corpus) of an AI model to influence or corrupt its learning process and outputs.

**Data Set**: A collection of data used by an AI model to learn and make decisions.

**Deep Learning (DL)**: A type of machine learning that uses artificial neural networks to process and learn from large amounts of data, similar to how the human brain works.

**Dropout**: A regularization technique where randomly selected neurons are ignored during training to prevent the model from over-relying on specific neurons and reduce overfitting.

**Embedding**: A method of representing discrete objects, like words or images, as continuous vectors in a lower-dimensional space to capture semantic meaning or relationships.

**Epoch**: A full cycle of training where the machine learning model processes the entire dataset once.

**F1 Score**: A metric that balances precision and recall to give an overall measure of a model’s accuracy.

**Feature**: An individual measurable property or characteristic used as input to an AI model, like height or color.

**Fine-Tuning**: The process of taking a pre-trained model and adjusting it for a new task by training it further on new data.

**Generative AI**: A type of AI that creates new content like text, images, or music based on the data it’s trained on.

**Gradient Descent**: An optimization algorithm used to minimize the loss function by iteratively adjusting the model's weights in the direction that reduces the error.

**Hallucination**: When a model generates content or information that seems real but is actually incorrect or fabricated, typically due to limitations in the model's training data.

**Hyperparameter**: Settings that control how an AI model learns, which are adjusted before training, like learning rate or number of layers.

**Inference**: The process of making predictions or decisions using a trained AI model on new data.

**Latent Space**: A lower-dimensional representation of data, where similar inputs are grouped together, often used in generative models like autoencoders.

**Learning Rate**: A hyperparameter that controls how much to change the model's weights with respect to the loss gradient during training.

**Loss Function**: A mathematical function that measures how far the model's predictions are from the actual values, used to guide the training process.

**Machine Learning (ML)**: A branch of AI where computers learn from data to make decisions or predictions without being explicitly programmed for every step.

**Model**: The mathematical structure or program that processes data to make predictions or decisions in AI.

**Natural Language Processing (NLP)**: The field of AI that focuses on teaching machines to understand and generate human language.

**Neural Network**: A series of algorithms modeled after aspects of the human brain, designed to recognize patterns and make decisions based on data.

**Neuron**: A basic unit in a neural network that takes in inputs, processes them, and sends an output to the next layer.

**Overfitting**: When a model learns the training data too well, including noise or irrelevant patterns, and performs poorly on new, unseen data.

**Precision**: The proportion of correct positive predictions out of all positive predictions made by a model.

**Prompt**: The input or instructions given to an AI model, often in the form of text, that guides the model's response or behavior.

**Prompt Injection**: A technique where malicious or unexpected input is provided to an AI system to manipulate its behavior or outputs, often used to exploit vulnerabilities in natural language models.

**Recall**: The proportion of actual positives that a model correctly identifies.

**Regularization**: Techniques used to prevent overfitting by discouraging the model from becoming too complex, often by adding a penalty for large weights.

**Reinforcement Learning**: A type of learning where a model improves its behavior based on rewards or punishments it receives for its actions.

**Supervised Learning**: A type of machine learning where the model is trained using labeled data (data with the correct answers provided).

**Training**: The process where an AI model learns from data by adjusting its internal settings to improve performance.

**Transformer**: A type of deep learning model architecture designed to process sequential data, like text, by focusing on the most relevant parts of the input using a mechanism called "attention."

**Underfitting**: When a model is too simple and fails to capture important patterns in the data, resulting in poor performance.

**Unsupervised Learning**: A type of machine learning where the model learns patterns and relationships in data without being given labeled answers.

**Vector**: A list or array of numbers that represents data in a form that a machine learning model can understand, often used to represent words, images, or other types of data.

**Weights**: Parameters in a machine learning model that determine how inputs are combined to produce the output, and are adjusted during training to improve the model's performance.
